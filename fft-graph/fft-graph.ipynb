# Welcome to the FFT Graphing Agent

```python
import json
import numpy as np
import matplotlib.pyplot as plt
from dataclasses import dataclass

filename = "readings.json"
```

## Ingesting JSON payload and ripping data
picking up a json and ingesting it

```python
f = open(filename,"r")
data = json.load(f)
print("file contents:")
print(data)

f.close()
```

Create collection of complex numbers

```python
complexNums = []

for element in data:
    newNum = complex(element["Real"], element["Imaginary"])
    complexNums.append(newNum)

print(f"# of items in array: {len(complexNums)}")
```

Start Processing to Create Frequency Signal.

```python
# sampling frequency
# To produce a 10-second wave
length = 10

# assumes sampling stage of 10 seconds
sample_rate = int(len(complexNums)/length)

# Casts the fourier transform coefficients as a numpy array
transformed_y = np.array(complexNums, dtype=complex)

# Take the absolute value of the complex numbers for magnitude spectrum
freq_magnitude = np.abs(transformed_y)

# Create frequency x-axis that will span up to sample_rate
freq_axis = np.linspace(0, sample_rate, len(freq_magnitude))

# Plot frequency domain
plt.plot(freq_axis, freq_magnitude)
plt.xlabel("Frequency (Hz)")
plt.xlim(0, 40)
plt.ylim(0, 10)
plt.show()
```

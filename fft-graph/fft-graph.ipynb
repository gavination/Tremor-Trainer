# Welcome to the FFT Graphing Agent

```python
import json
import complex_processor as processor
from dataclasses import dataclass

x_filename = "XReadingData.json"
y_filename = "YReadingData.json"
z_filename = "ZReadingData.json"
```

## Ingesting JSON payload and ripping data
picking up a json and ingesting it

```python
fx = open(x_filename,"r")
fy = open(y_filename, "r")
fz = open(y_filename, "r")

x_axis_data = json.load(fx)
y_axis_data = json.load(fy)
z_axis_data = json.load(fz)

fx.close()
fy.close()
fz.close()
```

Create collections of complex numbers

```python
x_complex_nums = processor.convert_to_complex_array(x_axis_data)
y_complex_nums = processor.convert_to_complex_array(y_axis_data)
z_complex_nums = processor.convert_to_complex_array(z_axis_data)
```

Start Processing to Create Frequency Signal.

```python
# To produce a 10-second wave
length = 10

processor.create_frequency_signal(x_complex_nums, length, "x values")
processor.create_frequency_signal(y_complex_nums, length, "y values")
processor.create_frequency_signal(z_complex_nums, length, "z values")
```
